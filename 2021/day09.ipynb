```python
# Next, you need to find the largest basins so you know what areas are most important to avoid.

# A basin is all locations that eventually flow downward to a single low point. Therefore, every low point has a basin, although some basins are very small. Locations of height 9 do not count as being in any basin, and all other locations will always be part of exactly one basin.
def get_points_flow_downward_to_xy(x,y, points):
    #points arround x,y
    xminus = lambda j: j-1 if j-1 >= 0 else j
    xplus = lambda j, m: j+1 if j+1 < m else j
    xl, xr = xminus(x), xplus(x,len(data))
    yl, yr = xminus(y), xplus(y,len(data[0]))
    for i in range(xl,xr+1):
        for j in range(yl,yr+1):
            if abs(i-x) + abs(j-y) == 1:
                #print(i,j,end='|')
                if data[i][j] !=9 and data[i][j] >= data[x][y] and (i,j) not in points:
                    points[(i,j)] = 0
    #print(points)
    return points

#get basin
def get_basins(x,y):
    points = {(x,y):1}
    points_new = get_points_flow_downward_to_xy(x,y, points.copy())
    while points_new != points:
        #print('while:', points_new, points)
        points = points_new.copy()
        #print('copy:', points_new, points)
        for(i,j) in points.keys():
            #print(i,j)
            if points[(i,j)] == 0 :
                points_new[(i,j)] = 1 
                points_new = get_points_flow_downward_to_xy(i,j, points_new.copy())
                #print('forres:', points_new, points)
    return len(points_new)

print(get_basins(4,6))
# The size of a basin is the number of locations within the basin, including the low point. The example above has four basins.

# The top-left basin, size 3:

# 2199943210
# 3987894921
# 9856789892
# 8767896789
# 9899965678
# The top-right basin, size 9:

# 2199943210
# 3987894921
# 9856789892
# 8767896789
# 9899965678
# The middle basin, size 14:

# 2199943210
# 3987894921
# 9856789892
# 8767896789
# 9899965678
# The bottom-right basin, size 9:

# 2199943210
# 3987894921
# 9856789892
# 8767896789
# 9899965678
# Find the three largest basins and multiply their sizes together. In the above example, this is 9 * 14 * 9 = 1134.

# What do you get if you multiply together the sizes of the three largest basins?
# get the three largest basins
def three_largest_basins(basins):
    basins = sorted(basins, reverse=True)
    print(basins[:3])
    return basins[0]*basins[1]*basins[2]
print(three_largest_basins(basins))
```
